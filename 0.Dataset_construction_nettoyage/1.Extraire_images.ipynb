```python
# définir la taille des parties en bytes
part_size = 1000000000

# fonction pour diviser le fichier en parties
def split_file(file_path, part_size):
    with open(file_path, 'rb') as f:
        # obtenir la taille totale du fichier
        file_size = os.path.getsize(file_path)
        # calculer le nombre de parties nécessaires
        num_parts = file_size // part_size + 1

        for i in range(num_parts):
            # obtenir le nom de fichier pour chaque partie
            part_name = f"{file_path}_{i}"
            # lire la partie du fichier
            data = f.read(part_size)
            # créer un nouveau fichier pour la partie
            with open(part_name, 'wb') as part_file:
                part_file.write(data)
```

```python
# fonction pour télécharger une partie précise
def download_part(file_path, part_num):
    # ouvrir le fichier de la partie
    part_file_path = f"{file_path}_{part_num}"
    with open(part_file_path, 'rb') as f:
        # lire les données de la partie
        data = f.read()
    return data
```

```python
image_urls = df2["image_ingredients_url"].tolist()
```

```python
image_urls
```

```python
# dossier où enregistrer les images téléchargées
save_dir = "/content/drive/MyDrive/OCR/Dataset/Images"
```

```python
# vérifier si le dossier de sauvegarde existe, sinon le créer
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
# parcourir la liste des URL d'images et télécharger chaque image
for i, image_url in enumerate(image_urls):
    if pd.notna(image_url):
        # effectuer une requête HTTP pour récupérer les données de l'image
        response = requests.get(image_url)

        # sauvegarder les données de l'image dans un fichier local
        filename = f"image_{i}.jpg"  # nom du fichier avec un numéro d'ordre
        filepath = os.path.join(save_dir, filename)  # chemin complet du fichier
        with open(filepath, "wb") as f:
            f.write(response.content)

        print(f"Téléchargement de l'image {i+1}/{len(image_urls)} terminé.")
    else:
        print(f"La ligne {i+1} ne contient pas d'URL d'image valide.")
```
